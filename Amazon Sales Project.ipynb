```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
amazon_sales = pd.read_csv("Amazon Sales data.csv")
amazon_sales.head()
```

```python
amazon_sales.info()
```

```python
amazon_sales.isnull().sum()
```

```python
amazon_sales.drop_duplicates()
```

```python
#Converting Order Date to datetime format
from datetime import datetime
amazon_sales["Order Date"]=pd.to_datetime(amazon_sales["Order Date"],format="%m/%d/%Y")
amazon_sales["Month"]=amazon_sales["Order Date"].dt.month
amazon_sales["Year"]=amazon_sales["Order Date"].dt.year
amazon_sales.head()
```

```python
# calculate the how many number of region,country,item types are there uniquely
regions=amazon_sales["Region"].nunique()
print("The Number of Regions =",regions)
countries=amazon_sales["Country"].nunique()
print("The Number of countries =",countries)
item_types=amazon_sales["Item Type"].nunique()
print("The Number of Item Types =",item_types)
```

```python
# calculate the total units sold,price,cost,revenue,profit,
units_sold = amazon_sales["Units Sold"].sum()
print("The Total Units sold are =",units_sold)
unit_price = amazon_sales["Unit Price"].sum()
print("The Total Unit Price is =",unit_price)
unit_cost=amazon_sales["Unit Cost"].sum()
print("The Totla Unit Cost is =",unit_cost)
total_revenue=amazon_sales["Total Revenue"].sum()
print("The Total Revenue is =",total_revenue)
total_cost=amazon_sales["Total Cost"].sum()
print("The Total Cost is =",total_cost)
total_profit=amazon_sales["Total Profit"].sum()
print("The Total Profits is =",total_profit)
```

```python
# Sales Across the Years
year_sales = amazon_sales.groupby(["Year"])["Total Revenue"].mean()
plt.figure(figsize=(10,4))
sns.barplot(x=year_sales.index,y=year_sales.values)
plt.title("Average Sales By Each Year")
plt.xlabel("Year")
plt.ylabel("Total Revenue")
plt.show()
plt.clf()
```

```python
# profits across the regions
region_profit = amazon_sales.groupby(['Region'])['Total Profit'].mean()
plt.figure(figsize=(4,4))
plt.pie(region_profit,labels=region_profit.index,autopct="%d%%")
plt.title('Average Profits Across Regions')
plt.show()
plt.clf()
```

```python
# total_revenue by each item type
plt.figure(figsize=(14,8))
ax=plt.subplot(1,1,1)
plt.bar(amazon_sales['Item Type'],amazon_sales['Total Revenue'],width=0.5,color='mediumseagreen')
ax.set_xticks(amazon_sales['Item Type'])
ax.set_xticklabels(amazon_sales['Item Type'])
plt.xticks(rotation= 25)
plt.xlabel('Item Type')
plt.ylabel("Total Revenue")
plt.yticks(np.arange(min(amazon_sales['Total Revenue']), max(amazon_sales['Total Revenue'])+10000, 1000000))
plt.title("Total Revenue by Each Item Type")
plt.show()
plt.clf()
```

```python
# which channel sales are producinh the high profit.
plt.figure(figsize=(6,6))
channel = amazon_sales.groupby(['Sales Channel'])['Total Profit'].mean()
myexplode = [0.01,0]
plt.pie(channel,labels=channel.index,autopct='%1.1f%%',startangle=90,explode=myexplode)
plt.title("The Total Profit by the Each Channel")
plt.show()
plt.clf()
```

```python
# Units sold across the regions
plt.figure(figsize=(14,8))
ax=plt.subplot(1,1,1)
plt.bar(amazon_sales['Region'],amazon_sales['Units Sold'],width=0.5,color='mediumseagreen')
ax.set_xticks(amazon_sales['Region'])
ax.set_xticklabels(amazon_sales['Region'])
plt.xticks(rotation= 25)
plt.xlabel('Regions')
plt.ylabel("Units Sold")
plt.yticks(np.arange(min(amazon_sales['Units Sold']), max(amazon_sales['Units Sold'])+1000,1000))
plt.title("Units Sold across Each Region")
plt.show()
plt.clf()
```

```python
# units sold and the profits.
plt.figure(figsize=(10,6))
ax=plt.subplot(1,1,1)
sns.regplot(x='Units Sold', y='Total Profit',data=amazon_sales,line_kws={'color':'darkslategray'},
            ci=None,scatter_kws={'s':50,'marker':'x','color':'teal'})
plt.title("Relation Ship Between Units Sold and Profit")
plt.yticks(np.arange(min(amazon_sales['Total Profit']), max(amazon_sales['Total Profit']),100000))
plt.grid(axis='y')
sns.despine()
plt.show()
plt.clf()
```

```python
# Profit Margin
profit_margin = (amazon_sales['Total Profit']/amazon_sales['Total Revenue'])*100
plt.figure(figsize=(8, 6)) 
plt.hist(profit_margin, rwidth=0.8, color='teal', bins=range(int(min(profit_margin)), (int(max(profit_margin)))+10,6))
plt.xlabel("Profit Margin(%)")
plt.ylabel("Frequnency")
plt.title("Profit Margin")
plt.show()
plt.clf()
```
